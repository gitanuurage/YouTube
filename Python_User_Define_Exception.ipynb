**User-defined exceptions in Python, also known as custom exceptions, are created to handle specific error conditions within a program that are not adequately covered by Python's built-in exceptions. These custom exceptions provide more clarity and granularity in error handling, making code more readable and maintainable.**

#**Basic User-Defined Exception**

# 1. Custom Exception Example

```python
class MyCustomError(Exception):
    pass

try:
    raise MyCustomError("This is a custom error message.")
except MyCustomError as mce:
    print('Caught  an exception: ', mce)
```

#**Examples with Specific Use Cases**

# 2. Age Restriction Exception

```python
class AgeTooLowError(Exception):
    pass

def check_age(age):
    if age < 18:
        raise AgeTooLowError("Age must be at least 18 to enter the club.")
    else:
        print('Age is valid.')

try:
    check_age(18)
except AgeTooLowError as atle:
    print(atle)
```

# 3. Negative Number Exception

```python
class NegativeNumberError(Exception):
    pass

def check_number(num):
    if num < 0:
        raise NegativeNumberError('Negative Number are not allowed.')
    else:
        print('going good', num)

try:
    check_number(10)
except NegativeNumberError as nne:
    print(nne)
```

# 4. Division by Zero Custom Exception

```python
class DivisionByZeroError(Exception):
    pass

def div_num(a, b):
    if b == 0:
        raise DivisionByZeroError("cannot divide by zero!! ")
    else:
        return a/b

try:
    div_num(12, 10)
except DivisionByZeroError as dbze:
    print(dbze)
```

# 5. Insufficient Balance Exception

```python
class InsufficientBalanceError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientBalanceError('Not enough balance in account to withdraw.')
    balance -= amount
    return balance

try:
    withdraw(1500, 1000)
except InsufficientBalanceError as ibe:
    print(ibe)
```

# 6. Invalid Email Format Exception

```python
class InvalidEmailError(Exception):
    pass

def check_emailid(emailid):
    if '@' not in emailid:
        raise InvalidEmailError('Invalid email address')
    else:
        print('valid email. Good to go.')

try:
    check_emailid('abcd@gmail.com')
except InvalidEmailError as iee:
    print(iee)
```

# 7. Invalid Password Exception

```python
class WeakPasswordError(Exception):
    pass

def validate_password(password):
    if len(password) < 6:
        raise WeakPasswordError("Password must be at least 6 characters long.")
    else:
        print('Good to go')

try:
    validate_password("123ABC")
except WeakPasswordError as e:
    print(e)
```

# 8. Out of Stock Exception

```python
class OutOfStockError(Exception):
    pass

def buy_item(stock, quantity):
    if quantity > stock:
        raise OutOfStockError("Item is out of stock.")
    stock -= quantity
    return stock

try:
    buy_item(15, 10)
except OutOfStockError as e:
    print(e)
```

# 9. Invalid Temperature Exception

```python
class TemperatureTooHighError(Exception):
    pass

def check_temperature(temp):
    if temp > 100:
        raise TemperatureTooHighError("Temperature is too high!")
    else:
        print('Good to go!!!')

try:
    check_temperature(50)
except TemperatureTooHighError as e:
    print(e)
```

# 10. Invalid Username Exception

```python
class InvalidUsernameError(Exception):
    pass

def check_username(username):
    if " " in username:
        raise InvalidUsernameError("Username should not contain spaces.")

try:
    check_username("john doe")
except InvalidUsernameError as e:
    print(e)
```
