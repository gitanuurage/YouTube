These exceptions are organized in a class hierarchy, with BaseException as the root, and Exception as the base class for most standard built-in exceptions.

# 1. ZeroDivisionError - Division by Zero

Raised when a number is divided by zero.

Raised when the second argument of a division or modulo operation is zero. The associated value is a string indicating the type of the operands and the operation.

```python
a = int(input('enter a number '))
b = int(input('enter a number '))

print(a/b)
```

```python
a = int(input('enter a number '))
b = int(input('enter a number '))

try:
    print(a/b)
except ZeroDivisionError as zde:
    print('Error: ', zde)
```

# 2. TypeError - Invalid Type Operation

Raised when an operation is performed on incompatible types.

Raised when an operation or function is applied to an object of inappropriate type. The associated value is a string giving details about the type mismatch.

This exception may be raised by user code to indicate that an attempted operation on an object is not supported, and is not meant to be. If an object is meant to support a given operation but has not yet provided an implementation, NotImplementedError is the proper exception to raise.

Passing arguments of the wrong type (e.g. passing a list when an int is expected) should result in a TypeError, but passing arguments with the wrong value (e.g. a number outside expected boundaries) should result in a ValueError.

```python
a = int(input('enter a number '))
b = input('enter a number ')

print(a + b)
```

```python
a = int(input('enter a number '))
b = input('enter a number ')

try:
    print(a + b)
except TypeError as te:
    print('Error: ', te)
```

# 3. ValueError - Invalid Value

Raised when an operation receives an invalid argument of the correct type.

Raised when an operation or function receives an argument that has the right type but an inappropriate value, and the situation is not described by a more precise exception such as IndexError.

```python
a = int('abc')
print(a)
```

```python
try:
    a = int('abc')
    print(a)
except ValueError as ve:
    print('Error: ', ve)
```

# 4. IndexError - List Index Out of Range

Raised when trying to access an index that does not exist in a list.

Raised when a sequence subscript is out of range. (Slice indices are silently truncated to fall in the allowed range; if an index is not an integer, TypeError is raised.)

```python
lst1 = [1,2,3,4,5]
print(lst1[5])
```

```python
lst1 = [1,2,3,4,5]
try:
    print(lst1[6])
except IndexError as ie:
    print('Error: ', ie)
```

# 5. KeyError - Dictionary Key Not Found

Raised when a dictionary key is not found.

Raised when a mapping (dictionary) key is not found in the set of existing keys.

```python
dict1 = {'a':'apple', 'b':'banana', 'c':'cherries', 'd':'dates'}
print(dict1['e'])
```

```python
dict1 = {'a':'apple', 'b':'banana', 'c':'cherries', 'd':'dates'}
try:
    print(dict1['e'])
except KeyError as ke:
    print('Error: ', ke)
```

# 6. AttributeError - Invalid Attribute Access

Raised when an attribute is accessed on an object that doesn’t have it.

Raised when an attribute reference (see Attribute references) or assignment fails. (When an object does not support attribute references or attribute assignments at all, TypeError is raised.)

```python
tuple1 = (0,1,2,3,4,5,6)

tuple1.append(7)
```

```python
tuple1 = (0,1,2,3,4,5,6)
try:
    tuple1.append(7)
except AttributeError as ae:
    print('Error: ', ae)
```

# 8. FileNotFoundError - File Does Not Exist

Raised when attempting to open a non-existing file.

Raised when a file or directory is requested but doesn’t exist.

```python
with open('non_existing_file.txt', 'r') as file:
    content = file.read()
```

```python
try:
    with open('non_existing_file.txt', 'r') as file:
        content = file.read()
except FileNotFoundError as fnfe:
    print('Error: ', fnfe)
```

# 9. ImportError - Module Import Failure

Raised when a module import fails.

Raised when the import statement has troubles trying to load a module. Also raised when the “from list” in from ... import has a name that cannot be found.

The optional name and path keyword-only arguments set the corresponding attributes:

name: The name of the module that was attempted to be imported.

path: The path to any file which triggered the exception.

```python
import non_existing_library
```

```python
try:
    import non_existing_library
except ModuleNotFoundError as mnfe:
    print('Error: ', mnfe)
```

```python
try:
    import non_existing_library
except ImportError as ie:
    print('Error: ', ie)
```

# 10. StopIteration - Iterator Exhausted

Raised when an iterator has no more items.

Raised by built-in function next() and an iterator's __next__() method to signal that there are no further items produced by the iterator.

value: The exception object has a single attribute value, which is given as an argument when constructing the exception, and defaults to None.

When a generator or coroutine function returns, a new StopIteration instance is raised, and the value returned by the function is used as the value parameter to the constructor of the exception.

If a generator code directly or indirectly raises StopIteration, it is converted into a RuntimeError (retaining the StopIteration as the new exception’s cause).

```python
myIter = iter([21,32,43,54,65,76])

print(next(myIter))
print(next(myIter))
print(next(myIter))
print(next(myIter))
print(next(myIter))
print(next(myIter))
print(next(myIter))
```

```python
myIter = iter([21,32,43,54,65,76])
try:
    print(next(myIter))
    print(next(myIter))
    print(next(myIter))
    print(next(myIter))
    print(next(myIter))
    print(next(myIter))
    print(next(myIter))
except StopIteration as si:
    print("Error Iterator Exhausted", si)
```
