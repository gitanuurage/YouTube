#**Basic Dictionary Comprehensions**

**1. Squares of Numbers**

```python
squares = {x : x**2 for x in range(1, 10)}
print(squares)
```

colab divider.png

**2. Mapping Numbers to Even/Odd**

```python
parity = {x : ('even' if x%2==0 else 'odd') for x in range(1, 10)}
print(parity)
```

colab divider.png

**3. Convert List to Dictionary**

```python
names = ['ram', 'shayam', 'shiv']
ages = [25, 30, 35]
person_dict = {names[i] : ages[i] for i in range(len(names))}
print(person_dict)
```

**4. Character Count in String**

```python
text = 'banana'
count = {ch : text.count(ch) for ch in set(text)}
print(count)
```

colab divider.png

**5. Filter Dictionary Values**

```python
num = {x : x**2 for x in range(10) if x%2==0 }
print(num)
```

colab divider.png

**6. Swap Keys and Values**

```python
my_dict = {"a":1, "b":2, "c":3}
swapped = {v: k for k, v in my_dict.items()}
print(swapped)
```

**7. ASCII Table for Characters**

```python
ascii_dict = {chr(i): i for i in range(65, 71)}
print(ascii_dict)   # {'A':65,'B':66,'C':67,'D':68,'E':69,'F':70}
```

**8. Words and Their Lengths**

```python
words = ["apple", "banana", "cherry"]
lengths = {w: len(w) for w in words}
print(lengths)
```

**9. Dictionary of Even Squares**

```python
even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_squares)
```

**10. First Letter as Key**

```python
fruits = ["apple", "apricot", "banana", "cherry"]
fruit_dict = {f[0]: f for f in fruits}
print(fruit_dict)
```
